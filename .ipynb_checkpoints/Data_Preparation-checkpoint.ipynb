```python
import os
import shutil
import glob
from tqdm import tqdm
```

```python
Raw_DIR = r'D:\Driver-Drowsiness-Detection-System\mrlEyes_2018_01\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]) :
        if i.split('_')[4] == '0':
            shutil.copy(src = dirpath + '/' + i, dst = r'D:\Driver-Drowsiness-Detection-System\mrlEyes_2018_01\Prepared_Data\Close_Eyes')
        
        elif i.split('_')[4] == '1':
            shutil.copy(src = dirpath + '/' + i, dst = r'D:\Driver-Drowsiness-Detection-System\mrlEyes_2018_01\Prepared_Data\Open_Eyes')
```
