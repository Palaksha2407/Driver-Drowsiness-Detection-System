```python
import os
import shutil
import glob
from tqdm import tqdm
```

```python
BASE_DIR = os.path.dirname(os.path.abspath('__file__'))
Raw_DIR = os.path.join(BASE_DIR, 'mrlEyes_2018_01', 'mrlEyes_2018_01')
Close_DIR = os.path.join(BASE_DIR, 'mrlEyes_2018_01', 'Prepared_Data', 'Close_Eyes')
Open_DIR = os.path.join(BASE_DIR, 'mrlEyes_2018_01', 'Prepared_Data', 'Open_Eyes')

# Create output directories if they don't exist
os.makedirs(Close_DIR, exist_ok=True)
os.makedirs(Open_DIR, exist_ok=True)

for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]) :
        if i.split('_')[4] == '0':
            shutil.copy(src = dirpath + '/' + i, dst = Close_DIR)
        
        elif i.split('_')[4] == '1':
            shutil.copy(src = dirpath + '/' + i, dst = Open_DIR)
```
